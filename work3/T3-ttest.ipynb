```python
from scipy.stats import ttest_1samp
scores = [75, 80, 68, 72, 77, 82, 81, 79, 70, 74, 76, 78, 81, 73, 81, 78, 75, 72, 74, 79, 78, 79]
mean = 75
alpha = 0.05

# t-test를 사용하여 가설 검정
t_statistic, p_value = ttest_1samp(scores, mean, alternative='greater')

print(t_statistic, p_value)

if p_value < alpha:
    print("귀무가설을 기각한다. 모평균은 75보다 크다")
else:
    print("귀무가설을 따르다. 모평균은 75이다.")
# alternative 파라미터의 사용법
# alternative='two-sided': 양측 검정 (기본값)
# alternative='greater': 단측 검정 (귀무가설이 클 경우)
# alternative='less': 단측 검정 (귀무가설이 작을 경우)
```
