```python
# 범주형 데이터 확인, Mode_of_Shipment, Product_importance

np.unique(df['Mode_of_Shipment']), np.unique(df['Product_importance']), np.unique(df['Gender'])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
cols = ['Mode_of_Shipment', 'Product_importance','Warehouse_block','Gender']
for col in cols:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('Reached.on.Time_Y.N', axis = 1)
y = df['Reached.on.Time_Y.N']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.head()

X_train = X_train.drop('ID', axis=1)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier


rf = RandomForestClassifier(n_estimators= 100, max_depth=5, random_state=1)
rf.fit(X_train, y_train)
```

```python
#test_id = X_test.pop('ID')
y_pred = rf.predict(X_test)
y_pred
```

```python
y_pred, y_test

np.sum(y_pred != y_test)
```

```python
print(f"total error: {np.sum(y_pred != y_test)} from {len(y_pred)}")
#len(y_pred
```
