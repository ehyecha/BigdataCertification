```python
#결측치 처리 


train_df.dropna(how='any', axis = 0, inplace=True)
test_df.dropna(how='any', axis = 0, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
#범주형 데이터
cols = ['Employment Type','GraduateOrNot','FrequentFlyer','EverTravelledAbroad']
le = LabelEncoder()
for col in cols:
    train_df[col] = le.fit_transform(train_df[col])
    test_df[col] = le.fit_transform(test_df[col])
```

```python
train_df.head()
```

```python
# train할 컬럼 선정 

y_train = train_df.pop('TravelInsurance')
```

```python
train_df
```

```python
#모델 선정 - classification

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 100, random_state= 42)
```

```python
rf.fit(train_df, y_train)
```

```python
y_pred = rf.predict_proba(test_df)
y_pred
```
